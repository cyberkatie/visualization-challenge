```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress

#data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

mouse_metadata
```

```python
study_results
```

```python
# Combine the data into a single dataset
merged = pd.merge(study_results, mouse_metadata,  on = 'Mouse ID', how = 'left')
# Display the data table for preview
merged.head()
```

```python
#find number of unique mice through Mouse ID
num_mice = merged['Mouse ID'].unique()
print(f'there are {len(num_mice)} mice in this study')
```

```python
#removes duplicates from the original dataframe where Mouse ID and Timepoint are the same
cleaned = merged.drop_duplicates(subset = ['Mouse ID', 'Timepoint'])
cleaned
```

# Summary Statistics

```python
#using groupby to calculate the variance and SEM
var = cleaned.groupby(['Drug Regimen']).var()['Tumor Volume (mm3)']
SEM = cleaned.groupby(['Drug Regimen']).sem()['Tumor Volume (mm3)']
```

```python
#using groupby and .describe() the find stats for each drug regimen
sorted_stats = cleaned.groupby(['Drug Regimen']).describe()['Tumor Volume (mm3)']
sorted_stats.drop(columns = ['count'], inplace = True)
sorted_stats
```

```python
#create a dataframe with the information we want
stats = {'Mean Tumor Volume' : sorted_stats['mean'], 'Median Tumor Volume' : sorted_stats['50%'],
        'Tumor Volume Variance' : var, 'Tumor Volume Std. Dev.' : sorted_stats['std'],
        'Tumor Volume Standard Error' : SEM}

stats_df = pd.DataFrame(stats)
stats_df
```

```python
#modified dataframe
summ_stats = {'Mean' : sorted_stats['mean'], 'Median' : sorted_stats['50%'], 'Variance' : var,
             'STD' : sorted_stats['std'], 'SEM' : SEM}
summary_stats_df = pd.DataFrame(summ_stats)
summary_stats_df
```

# Quartiles, Outliers, and Boxplots


```python
#sorting mice by drug regimen
cap_mice = cleaned[(cleaned['Drug Regimen'] == 'Capomulin')]
ram_mice = cleaned[cleaned['Drug Regimen'] == 'Ramicane']
inf_mice = cleaned[cleaned['Drug Regimen'] == 'Infubinol']
cef_mice = cleaned[cleaned['Drug Regimen'] == 'Ceftamin']

#finding the last timepoint for each mouse in each regimen
final_times_cap = cap_mice.groupby(['Mouse ID']).last()['Timepoint']
final_times_ram = ram_mice.groupby(['Mouse ID']).last()['Timepoint']
final_times_inf = inf_mice.groupby(['Mouse ID']).last()['Timepoint']
final_times_cef = cef_mice.groupby(['Mouse ID']).last()['Timepoint']
```

```python
#using for loops to find the final tumor volumes for each mouse in each regimen
final_vol_cap = []
final_vol_ram = []
final_vol_inf = []
final_vol_cef = []
for i in range(len(final_times_cap)):
    final_vol_cap.append(cap_mice['Tumor Volume (mm3)'].where(cap_mice['Timepoint'] == final_times_cap[i]))
    
for i in range(len(final_times_ram)):
    final_vol_ram.append(ram_mice['Tumor Volume (mm3)'].where(ram_mice['Timepoint'] == final_times_ram[i]))
    
for i in range(len(final_times_inf)):
    final_vol_inf.append(inf_mice['Tumor Volume (mm3)'].where(inf_mice['Timepoint'] == final_times_inf[i]))
    
for i in range(len(final_times_cef)):
    final_vol_cef.append(cef_mice['Tumor Volume (mm3)'].where(cef_mice['Timepoint'] == final_times_cef[i]))
```

```python
final_vol_cap = final_vol_cap[0]
final_vol_ram = final_vol_ram[0]
final_vol_inf = final_vol_inf[0]
final_vol_cef = final_vol_cef[0]
```

```python
#drop NaN values
final_vol_cap.dropna(inplace = True)
final_vol_ram.dropna(inplace = True)
final_vol_inf.dropna(inplace = True)
final_vol_cef.dropna(inplace = True)
```

```python
data = {'Capomulin' : final_vol_cap,
       'Ramicane' : final_vol_ram,
       'Infubinol' : final_vol_inf,
       'Ceftamin' : final_vol_cef}
drug_treatments_final = pd.DataFrame(data)
```

# Correlation and Regression

```python
xvalues = cap_mice['Weight (g)']
yvalues = cap_mice['Tumor Volume (mm3)']
```

```python
print(f"The correlation coefficient between mouse weight and tumor size is {round(st.pearsonr(xvalues, yvalues)[0],2)}")
```

```python
#linear regression
(slope, intercept, rvalue, pvalue, stderr) = linregress(xvalues, yvalues)
```

```python
regress_values = xvalues * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
```

```python
plt.scatter(xvalues,yvalues)
plt.plot(xvalues,regress_values,"r-")
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Size (mm3)')
plt.show()
```
